```python
name=[]
price=[]
rating=[]
description=[]
discount=[]
for j in range(1,11):
    url="https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}".format(j)
    req=requests.get(url)
    soup=BeautifulSoup(req.text,"html.parser") 
    result=soup.find_all("div",attrs={"class":"_2kHMtA"})
    for i in result:
    #product name
        name.append(i.find("div",attrs={"class":"_4rR01T"}).text)
    #price
        if (i.find("div",attrs={"class":"_30jeq3 _1_WHN1"})==None):
            price.append(" ")
        else:
            price.append(i.find("div",attrs={"class":"_30jeq3 _1_WHN1"}).text) 
    #rating
        if (i.find("div",attrs={"class":"_3LWZlK"})==None): 
            rating.append(" ")
        else:
            rating.append(i.find("div",attrs={"class":"_3LWZlK"}).text)
            
    #description
        description.append(i.find("li",attrs={"class":"rgWa7D"}).text )
    #discout
        if (i.find("div",attrs={"class":"_3Ay6Sb"})==None):
           discount.append("0% off")
        else:
            discount.append(i.find("div",attrs={"class":"_3Ay6Sb"}).text)
```

```python
name
```

```python
prod_detail={"Product Name":name,"Product Price":price,"Rating":rating,"Description":description,"Discount":discount}
```
